# üìò Impact des d√©penses publiques sur la performance scolaire aux √âtats-Unis (1990‚Äì2019)

```python
#importation des biblioth√®ques n√©cessaires
import pandas as pd # biblioth√®que essentielle pour la manipulation des donn√©es en tableau
import numpy as np # pour les calculs num√©riques et manipulation des matrices
import matplotlib.pyplot as plt # pour tracer des graphiques
import seaborn as sns # pour des visualisatiojns plus jolies et statistiques

#Biblioth√®ques pour le machine learning
from sklearn.model_selection import  train_test_split, cross_val_score, TimeSeriesSplit
#train_test_split: s√©parer nos donn√©es en train/test
#cross_val_score: √©valuer les modeles avec validation crois√©e
#TimeSeriesSplit: d√©coupage sp√©cifique pour s√©ries temporelles

from sklearn.linear_model import LinearRegression,Ridge
#LinearRegression: mod√®le de r√©gression lineaire classique
#Ridge: r√©gression lineaire avec r√©gulairsation pour √©viter le sur-apprentissage
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
#RandomFrestRegressor: un ensemble d'arbre de decision(methode)
#GradientBoostingRegressor: une m√©thode boosting qui combine plusieurs modeles faibles

from sklearn.metrics import mean_squared_error, r2_score
#mean_squared_error: mesure de l'erreur quadriatique moyenne
#r2_score: indique la qualit√© de la prediction(proche de 1  = tr√®s bon mod√®le)
```

## 1. Chargement et filtrage du dataset (1990‚Äì2019)

```python
#chargement et pr√©paration des donn√©es
dt = pd.read_csv("states_all.csv")
```

```python
#s√©lection des colonnes utiles
cols = [ "YEAR","STATE","ENROLL", "TOTAL_REVENUE",
       "FEDERAL_REVENUE", "STATE_REVENUE", "LOCAL_REVENUE",
       "TOTAL_EXPENDITURE", "INSTRUCTION_EXPENDITURE",
       "SUPPORT_SERVICES_EXPENDITURE","AVG_MATH_8_SCORE","AVG_READING_8_SCORE"
       ]
```

```python
#supprimer les lignes manquantes(NAN) et garder uniquement les colonnes qui nous int√©ressent
df = dt[cols].dropna()
```

```python
#on  filtre pour ne que la p√©riode de 1990-2019
df = df[(df["YEAR"] >= 1990) & (df["YEAR"] <=2019)]
```

```python
#cr√©ton de nouvelles variables : d√©penses par √©l√®ve
for col in ["TOTAL_EXPENDITURE","INSTRUCTION_EXPENDITURE","SUPPORT_SERVICES_EXPENDITURE",
            "FEDERAL_REVENUE","STATE_REVENUE","LOCAL_REVENUE"]:
    df[col + "_PER_STUDENT"] = df[col] / df["ENROLL"]
    #Exemple: TOTAL_EXPENDITURE_PER_STUDENT = TOTAL_EXPENDITURE / NOMBRE D'ELEVES

print("Taille apr√®s nettoyage:", df.shape)
#on affiche la taille du dataframe dinal apr√®s nettoyage pour v√©rifier

df.head(10)
```

## 3. Pr√©paration des donn√©es pour la mod√©lisation

```python
#Pr√©paration des donn√©es puor la modelisation
#information
#Classification(pr√©dire une cat√©gorie)
#variable explicative : √¢ge,revenu,nombre d'achats
#variable cible: "client fid√®le" (Oui/Non)

#Regression (predire une valeur num√©rique)
#variable explicative: surface ,quartier,nombre de chambre,ann√©e de contruction
#variable cible: prix de la maison, en $



#pr√©paration des variables explicatives (features) et cibles (targets)
features = [c for c in df.columns if "PER_STUDENT" in c]
#features: liste de noms de colonnes contenant "PER_STUDENT"
#df.columns: liste des colonnes de notre dataframe df
#c for in df.columns: on parcourty chaque colonne et on l'appel c
#[c for in df.columns if "PER_STUDENT" in c]: construit une contenant des conlonnes filtr√©es


#on d√©fenit deux cibles: le score moyen en math√©matique et score moyen en lecture
target_math ="AVG_MATH_8_SCORE"
target_read = "AVG_READING_8_SCORE"

#definition des variables en machine learning
X = df[features]
y_math = df[target_math]
y_read = df[target_read]
#X contient les varibales explicatices(d√©penses par √©l√®ve)
#y_math et y_read contiennet les varibles cibles ou √† pr√©dire(les scores)


#S√©paration en jeu d'entra√Ænement et de test bas√© sur temps
train_idx = df["YEAR"] <=2014
test_idx = df["YEAR"] >= 2015

#On entra√Æne le mod√®le sur les ann√©es jusqu'√† 2014 et on test sur 2015-2019
X_train,X_test = X[train_idx], X[test_idx]
y_math_train,y_math_test = y_math[train_idx],y_math[test_idx]
y__read_train,y_read_test = y_read[train_idx],y_read[test_idx]
# X[train_idx]: les lignes de X pour l'entra√Ænement(avant 2015)
#X[test_idx]: les lignes de X pour le test (en 2015 ET apr√®s)
#y_math[train_idx]: les scores de maths pour l'entrainemet 
#y_math[test_idx]: les scores de maths pour le test
# y_read[train_idx],y_read[test_idx]: m√™me logique pour la lec#verification de dimension de l'ensembles d'entra√Ænemeture


#verification de dimension de l'ensembles d'entra√Ænement et de test

X_train.shape, X_test.shape
```

## 4. Entra√Ænement et comparaison des mod√®les

```python
#Entra√Æneemnt et comparaison des mod√®les de machine learning
models = {
    "LinearRegression": LinearRegression(),
    #LinearRegression(): il suppose que la relation entre X et y est lin√©aire.
    #avantage: facile √† interpr√©ter.
    #Incov√©nient: trop limit√© si la relation n'est pas lin√©aire
    
    "Ridge": Ridge(alpha=1.0),
    #Rdge: est une regression regularis√©e.
    #nous permet d'√©viter le sur-apprentissage (Overfitting) en ajoutant une p√©nalit√© sur les co√©fficient trop grands.
    #ici alpha= 1.0 contr√¥le la force de la r√©gularistion. Plus alpha est grand, plus on p√©nalise les gros co√©fficients
    #le mod√®le devient plus simple, mais peut perdre en pr√©cision.
    
    "RandomForest": RandomForestRegressor(n_estimators=100, random_state=42),
    #c'est un mod√®le d'ensemble bas√© sur des arbres de d√©cision.
    #avantage: tr√®s robuste, capture les relations non lin√©aires, moins sensible au sur-apprentissage que les arbres simple.
    #n_estimators=100: Nombre d'abre dans la for√™t
    #random_state=42: fixe l'al√©atoire pour rendre les r√©sultats reproductibles
    #Un abre d√©cision en machine learning est un mod√®le predictif qui des d√©cisions suivant une structure arborescente(comme un orgranigramme).
    #on l'utlise aussi bien pour des t√¢ches de classification(pr√©dire les cats) que r√©gression(pr√©dire une valeur num√©rique)

    "GradientBoosting": GradientBoostingRegressor(n_estimators=100, random_state=42)
    #GrandientBoostingRegressor: un autre mod√®le d'ensemble, mais diff√©rent de la for√™t al√©atoire
    #n_estimators=100: Nombre d'abre dans la for√™t
    #random_state=42: fixe l'al√©atoire pour rendre les r√©sultats reproductibles

}


#On cr√©e un dictionnaire vide pour stocker les performances de chaque mod√®le.
#Chaque mod√®le sera identifi√© par son nom et associ√© √† ses scores (MSE, R¬≤).

results = {}

for name, model in models.items():
    model.fit(X_train, y_math_train)
    # On entra√Æne le mod√®le sur les donn√©es d‚Äôentra√Ænement (pr√©diction des scores de math).
    
    y_pred = model.predict(X_test)
    # On g√©n√®re des pr√©dictions sur la p√©riode de test (2015-2019).
  
    
#On va √©valuer la qualit√© de chaque mod√®le en utilisant deux m√©triques classiques en r√©gression :
#MSE (Mean Squared Error) : erreur quadratique moyenne. Plus c‚Äôest petit ‚Üí mieux c‚Äôest.
#R¬≤ (Coefficient de d√©termination) : indique combien de variance du vrai signal est expliqu√©e par le mod√®le. Plus proche de 1 ‚Üí meilleur mod√®le.
    mse = mean_squared_error(y_math_test, y_pred)
    r2 = r2_score(y_math_test, y_pred)
    # On calcule deux m√©triques :
    # - MSE : plus il est petit, meilleure est la pr√©cision.
    # - R¬≤ : plus il est proche de 1, meilleure est la qualit√© du mod√®le.
    
    results[name] = {"MSE": mse, "R¬≤": r2}
    # On stocke les r√©sultats pour comparaison.

pd.DataFrame(results).T
# On affiche les r√©sultats sous forme de tableau.
```

## 5. Visualisation des r√©sultats (Math 8)

```python
#Visualisatin des r√©sultats (Math 8)
best_model =RandomForestRegressor(n_estimators=100, random_state=42)
best_model.fit(X_train,y_math_train)
y_pred = best_model.predict(X_test)

#RandomForestRegressor(n_estimators=100, random_state=42): cr√©er un mod√®le r√©gression bas√© sur une for√™t de donn√©e al√©atoire(ensemble d'abre de d√©cision)
#fit(X_train,y_math_train): entra√Æne le mod√®le avec les donn√©es d'apprentissage( caract√©ristiques X_train, et les resultat y_math_train)
#y_pred: permet de visualiser la qualit√© des pr√©dictions (si les points sont proches de la diagonale, √ßa veut dire que le mod√®le est bon)
#.predict(X_test): nous permet de faire la pr√©diction sur les donn√©es de test

#tracer un nuage de point entre les valeurs r√©elles (y_math_test) et pr√©dictions

plt.figure(figsize=(8,6))
plt.scatter(y_math_test,y_pred, alpha=0.7)
#alpha=0.7: cela signifie que chaque point est l√©g√®rement transparent (70% opaque, 30% transparent)
plt.xlabel("Valeurs r√©elles")
plt.ylabel("Pr√©dictions")
plt.title("Pr√©diction du score en Math√©matique (Grade 8)")
plt.show()
```

## 6. Importance des variables

```python
#Importance des variables sur notre Mod√®le

importances = best_model.feature_importances_
feat_imp = pd.Series(importances,index=features).sort_values(ascending=False)
#importances = best_model.feature_importances_: on r√©cup√®re l'importance de chaque variable(feature) dans le mod√®le best_model
#feat_import = pd.Series(importances,index=features).sort_values(ascending=False): On cr√©e une serie Pandas o√π les valeurs sont importance des variables
# et les index sont les noms des variables(features) et on trie les variables de la plus importante √† la moins importante

plt.figure(figsize=(8,5))
sns.barplot(x=feat_import,y=feat_imp.index)
plt.title("Importance des varibles en Math√©matique niveau 8 √®me ann√©e")
plt.show()
```
